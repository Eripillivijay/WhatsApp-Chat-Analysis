#### Preprocessing Data

```python
import re
import pandas as pd
```

```python
f = open('chat/satya.txt','r',encoding='utf-8')
```

```python
data = f.read()
```

```python
print(type(data))
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{4},\s\d{1,2}:\d{2}\s*[ap]m\s-\s'
```

```python
messages = re.split(pattern, data)[1:]
messages
```

```python
dates = re.findall(pattern, data)
dates
```

```python
df = pd.DataFrame({
    'user_message': messages,
    'message_date': dates
    })

# Converting Message_Data type
df['message_date'] = df['message_date'].str.upper()
df['message_date'] = pd.to_datetime(
    df['message_date'], 
    format='%d/%m/%Y, %I:%M %p - '
    )

df.rename(
    columns={'message_date': 'date' },
    inplace=True
)
```

```python
df.head()
```

```python
df.shape
```

```python
# separating the user-name and user-data

users = []
messages = []

for message in df['user_message']:
    entry = re.split(r'([\w\W]+?):\s', message)
    if entry[1:]:  # user name
        users.append(entry[1])
        messages.append(" ".join(entry[2:]))
    else:
        users.append('group_notification')
        messages.append(entry[0])

df['user'] = users
df['message'] = messages
df.drop(columns=['user_message'], inplace=True)
```

```python
df.head(1)
```

```python
df['only_date'] = df['date'].dt.date
df['year'] = df['date'].dt.year
df['month_num'] = df['date'].dt.month
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['day_name'] = df['date'].dt.day_name()
df['hour'] = df['date'].dt.hour
df['minute'] = df['date'].dt.minute
```
